```python
import pandas as pd
import os
```

```python
fpl_df = pd.read_csv("./Data/gw_history_['@FPL Matthew'].csv")

files = [file for file in os.listdir('./Data')]

all_managers = pd.DataFrame()

for file in files:
   df = pd.read_csv("./Data/"+file)
   df['manager'] = file[12:-5]
   all_managers = pd.concat([all_managers,df])

all_managers.to_csv('gw_history_raw.csv', index=False)
```
